```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../Data/EV sales Q1-2022-catagory wise.csv')
df.head(20)
```

```python
df = df.sample(frac=1.0, random_state=42)
df = df.reset_index(drop=True)
```

```python
def convert_to_integer(value_with_commas):
    value_with_commas = value_with_commas.replace(",", "")
    cleaned_value = value_with_commas.replace(" ", "")
    return int(cleaned_value)
columns_to_convert = ['Jan - 22', 'Feb - 22', 'Mar - 22' ,'Apr - 22', 'May - 22', 'Jun - 22', 'Total']
df[columns_to_convert] = df[columns_to_convert].applymap(convert_to_integer)
```

```python
df.head(10)
```

```python
df['Total'] = df['Jan - 22'] + df['Feb - 22'] + df['Mar - 22'] + df['Apr - 22'] + df['May - 22'] + df['Jun - 22']
```

```python
plt.figure(figsize=(12, 8))
df_melted = pd.melt(df, id_vars=['State'], value_vars=['Jan - 22', 'Feb - 22', 'Mar - 22', 'Apr - 22', 'May - 22', 'Jun - 22'],
                     var_name='Month', value_name='Sales')

# Define a custom color palette for the lines
custom_palette = sns.color_palette("husl", n_colors=len(df['State'].unique()))

# Create a line plot to compare monthly sales for different states
plt.figure(figsize=(12, 12))
sns.lineplot(data=df_melted, x='Month', y='Sales', hue='State' , marker='o')

# Add labels and a title
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Comparison by State')

# Rotate x-labels for better readability
plt.xticks(rotation=45)

# Show legend
plt.legend(title='State')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Group the data by state and calculate the total sales for each state
state_totals = df.groupby('State')['Total'].sum().reset_index()

# Create a 2D pie chart
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(state_totals['Total'], labels=None, autopct='%1.1f%%', startangle=140)

# Add labels and percentages to the legend
labels = state_totals['State']
percentages = ["{:.1f}%".format(pct) for pct in (state_totals['Total'] / state_totals['Total'].sum() * 100)]
legend_labels = [f'{label} ({percentage})' for label, percentage in zip(labels, percentages)]
plt.legend(wedges, legend_labels, title="States", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

# Add a title
plt.title('Distribution of Total Sales by State')


# Show the plot
plt.axis('equal')
plt.show()
```

```python
X = df[['Category','Jan - 22', 'Feb - 22', 'Mar - 22', 'Apr - 22', 'May - 22', 'Jun - 22','Total' ]]
from sklearn.preprocessing import LabelEncoder
# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Apply label encoding to the 'Category' column
X['Category'] = label_encoder.fit_transform(X['Category'])
df['State'] = label_encoder.fit_transform(df['State'])
df['Category'] = label_encoder.fit_transform(df['Category'])

```

```python
from sklearn.manifold import TSNE
# Assuming 'X' is your high-dimensional data
tsne = TSNE(n_components=2, random_state=42)  # You can change the number of components (e.g., n_components=3)
X_tsne = tsne.fit_transform(X)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=df['State'], cmap='viridis')  # 'your_labels' are labels for coloring data points
plt.colorbar()
plt.title('t-SNE Visualization')
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Standardize the data (important for PCA)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(X)

# Apply PCA
pca = PCA(n_components=2)  # Specify the number of components to keep (e.g., 2)
pca_result = pca.fit_transform(scaled_data)

# Create a DataFrame for the PCA results
pca_df = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2'])

# Display the PCA DataFrame
print(pca_df)

# Variance explained by each component
explained_variance = pca.explained_variance_ratio_
print("Explained Variance Ratios:", explained_variance)

# Scree plot (explained variance)
plt.bar(range(1, len(explained_variance) + 1), explained_variance)
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot')
plt.show()
```

```python
from sklearn.cluster import KMeans
# Select the features you want to use for clustering (e.g., 'PC1' and 'PC2')
X = pca_df[['PC1', 'PC2']]
```

```python
# Example: Using the elbow method to find an appropriate K value
wcss = []  # Within-cluster sum of squares
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# Plot the elbow curve to visualize an appropriate K
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=2, init='k-means++', max_iter=300, n_init=10, random_state=0)
kmeans.fit(X)
```

```python
pca_df['Cluster'] = kmeans.labels_
```

```python
# Scatter plot with different colors for each cluster
plt.scatter(X['PC1'], X['PC2'], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, c='black', label='Centroids')
plt.title('K-means Clustering')
plt.xlabel('Feature1')
plt.ylabel('Feature2')
plt.legend()
plt.show()
```
